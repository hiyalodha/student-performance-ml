```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/cleaned_student_data.csv")

df.head()
df.describe()
```

```python
df[["math_score", "reading_score", "writing_score"]].hist(
    bins=20,
    figsize=(10, 6)
)
plt.suptitle("Distribution of Student Scores")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.heatmap(
    df[["math_score", "reading_score", "writing_score"]].corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Between Scores")
plt.show()
```

```python
#EDA Insights
# - The average math score is around 66, while reading and writing scores are slightly higher.
# - Reading and writing scores show a very strong positive correlation (~0.95).
# - Math score also correlates positively with reading (~0.82) and writing (~0.80).
# - This suggests that students who perform well in reading and writing tend to also perform well in math.
# - No extreme outliers are observed in score distributions.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.boxplot(x="gender", y="math_score", data=df)
plt.title("Math Score Distribution by Gender")
plt.show()
```

```python
# Key EDA Insights
#- Reading and writing scores have a very strong positive correlation.
#- Math scores are slightly lower on average compared to reading and writing.
#- Male students show a slightly higher median math score than female students.
#- Score distributions are approximately normal with a few low-score outliers.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.boxplot(x="test_preparation_course", y="math_score", data=df)
plt.title("Math Score vs Test Preparation Course")
plt.xlabel("Test Preparation Course")
plt.ylabel("Math Score")
plt.show()
```

```python
#EDA Insights
#- Math, reading, and writing scores are strongly correlated with each other.
#- Students who completed the test preparation course tend to score higher in math.
#- Male students show slightly higher median math scores compared to female students.
#- Score distributions are approximately normal with a few low-score outliers.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/cleaned_student_data.csv")

df.head()
```

## Key Insights from Exploratory Data Analysis

1. **Score Distribution**
   - Math, reading, and writing scores are approximately normally distributed.
   - Most students score between 6080 across subjects.

2. **Correlation Between Scores**
   - Reading and writing scores show a very strong positive correlation.
   - Math score also correlates positively with reading and writing, but slightly less.

3. **Gender vs Math Score**
   - Male students show a slightly higher median math score compared to female students.
   - However, score distributions overlap significantly.

4. **Test Preparation Course**
   - Students who completed the test preparation course tend to score higher in math.
   - This suggests preparation has a positive impact on performance.
